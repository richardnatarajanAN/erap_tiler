```python
def find_occurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]

def replace_char(s, i, c):
    return s[:i] + c + s[i+1:]

find_occurrences('AATAT', 'T')
```

```python
def extract_features(seq: str):
    seq_sr = str_to_fasta_seqrecord('seq', seq)
    ref_sr = str_to_fasta_seqrecord('ref', erap_reference)

    ali = parasailer.run(ref_sr, seq_sr)

    query, ref = ali.traceback.query, ali.traceback.ref

    boundary_idx = find_occurrences(ref, '|')
    for i in boundary_idx:
        query = replace_char(query, i, '|')

    return clean_sequence(query, '-*.').split('|')

def print_ali(ref, query):
    
    if len(query) != len(ref):
        raise Exception('not equal lens?')
    ali_len = len(ref)
    print(ali_len)
    for i in range(0, ali_len, 80):
        print(ref[i:i+80], query[i:i+80], sep='\n', end='\n\n')
```
