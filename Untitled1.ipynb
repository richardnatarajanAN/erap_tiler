```python
file = 'tmp/ali.fasta'

from tgsts.sequtils.fastio import read_fasta

alis = read_fasta(file)[0:2]
```

```python
ali_len = len(alis[0])
```

```python
ali_strs = {
    str(ali.seq)
    for ali
    in alis
}
len(ali_strs)
```

```python
for i in range(0, ali_len, 80):
    s = slice(i,i+80)

    ali_snippets = [
        str(a[s].seq)
        for a
        in alis
    ]

    mm_line = ''.join([
        'v' if len(set(bases)) != 1 else ' '
        for bases 
        in zip(*ali_snippets)
    ])
    print(mm_line)
    print(*ali_snippets, sep='\n', end='\n\n')
```
